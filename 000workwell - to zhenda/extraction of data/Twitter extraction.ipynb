ComfortDelgro

```python
import snscrape.modules.twitter as sntwitter
import pandas

# Creating list to append tweet data to
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('comfortdelgro since:2021-01-18 until:2021-04-18').get_items()):
    if i>800:
        break
    tweets_list2.append([tweet.date, tweet.id, tweet.content, tweet.username])
    
# Creating a dataframe from the tweets list above
comfortdelgro_df = pd.DataFrame(tweets_list2, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
```

```python
comfortdelgro_df
```

```python
import numpy as np
np.unique(comfortdelgro_df['Username'])
```

```python
user_list = ['BusinessTimes','ChannelNewsAsia','TODAYonline','IndependentSG', 'techinasia', 'LinezNews', 'readtheedge_sg', 'MSN_Singapore']
comfortdelgro_df = comfortdelgro_df[~comfortdelgro_df['Username'].isin(user_list)]
```

```python
comfortdelgro_df
```

```python
comfortdelgro_df['Username'].unique()
```

```python
comfortdelgro_df.to_csv(path_or_buf = 'C:/SMU/Y3S2/FYP/NEW_DATA/comfort_delgro.csv', index = False)
```

Grab

```python
# Creating list to append tweet data to
tweets_list3 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('grabsg since:2018-01-18 until:2020-01-18').get_items()):
    if i>800:
        break
    tweets_list3.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.user.followersCount])
    
# Creating a dataframe from the tweets list above
grab_df = pd.DataFrame(tweets_list3, columns=['Datetime', 'Tweet Id', 'Text', 'Username', 'Followers'])
```

```python
grab_df
```

```python
grab_df = grab_df[grab_df['Followers'] < 100]
```

```python
grab_df.to_csv(path_or_buf = 'D:/Y3S2/FYP/grab_tweets.csv', index = False)
```
